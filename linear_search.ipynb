```python
def linear_search(arr,target):
  size = len(arr)
  for i in range(0,size+1):
    if arr[i] == target:
      return i
  return -1
```

```python
lst = [10,2,3,56,20]
target = 3

result = linear_search(lst,target)
print(result)
```
