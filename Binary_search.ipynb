```python
def B_search(arr,target):
  size = len(arr)
  start = 0
  end = size-1
  for i in range(start,end):
    mid = (start+end)//2
    if  arr[mid]==target:
      return mid
    elif arr[mid]<target:
      start = mid+1
    elif arr[mid]>target:
      end = mid-1
  return -1
```

```python
sorted_array = [1,4,6,7,12,56,78,85,90]
target = 56
result = B_search(sorted_array,target)
print(f"result is : {result}")
```
