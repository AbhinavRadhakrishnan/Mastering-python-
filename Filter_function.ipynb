Filter function is used to filter out items from a list based on a specific condition

```python
def even(num):
  if num%2==0:
    return True
even(24)
```

```python
lst =[1,2,3,4,5,6]
evn =[]
for num in lst:
  if num%2==0:
    evn.append(num)
print(evn)
```

```python
#by using filter
list(filter(even,lst))
```

```python
#filter with lambda function
num=[1,2,3,4,5,6,7,8,9]
greater_than_five = list(filter(lambda x:x>5,num))
print(greater_than_five)
```

```python
#filter with a lambda function with multiple conditions
num
```

```python
even_and_greater_than_five = list(filter(lambda x:x>5 and x%2==0,num))
print(even_and_greater_than_five)
```

```python
#filter to check if is greater than 25 in dict

people = [
    {'name':'john','age':20},
    {'name':'jonny','age':27},
    {'name':'john','age':29},
]
```

```python
def age_greater_than_25(person):
  return person['age']>25
```

```python
list(filter(age_greater_than_25,people))
```
