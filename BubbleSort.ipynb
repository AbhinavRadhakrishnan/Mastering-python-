```python
def bubble_sort(lst):
  size = len(lst)
  for passes in range(0,size-1):

    for i in range(0,size-1-passes):
     if lst[i] > lst[i+1]:
      lst[i],lst[i+1]=lst[i+1],lst[i]
  return lst

```

```python
unsorted_list= [5,4,3,2,1]
sorted_list =bubble_sort(unsorted_list)
print(sorted_list)
```
